## Karar Ağaçları (Decision Trees)

```python
# https://erdincuzun.com/makine_ogrenmesi/decision-tree-karar-agaci-id3-algoritmasi-classification-siniflama/
# https://en.wikipedia.org/wiki/Decision_tree
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Veri oku
# https://rdrr.io/cran/rpart/man/kyphosis.html
df = pd.read_csv('https://raw.githubusercontent.com/yasarkucukefe/YBS462/main/data/kyphosis.csv')
```

```python
df.head()
```

```python
# null değerler?
df.info()
```

```python
# EDA
sns.pairplot(df, hue='Kyphosis')
```

```python
# Train-test split (eğitim ve test verilerinin ayrılması)
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# Decision Trees Model (Karar Ağaçları Modeli)
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
# Tahminler
tahminler = dtree.predict(X_test)
```

```python
tahminler
```

```python
# Modelin değerlendirilmesi
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, tahminler))
```

```python
y_test[y_test=='absent'].size / y_test.size
```

```python
print(confusion_matrix(y_test, tahminler))
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(18,12))
tree.plot_tree(dtree, fontsize=10, feature_names=list(df.columns[1:]))
plt.show()
```

## Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50) # n_estimators=100 (default)
```

```python
rf.fit(X_train, y_train)
```

```python
rf_tahminler = rf.predict(X_test)
```

print(confusion_matrix(y_test, rf_tahminler))
