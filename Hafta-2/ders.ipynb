# Grafikler (matplotlib)

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
print(f"x={x}, y={y}")
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r')
plt.show()
```

```python
y2 = x ** 3
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y2,'g*-')
```

## Seaborn

```python
import seaborn as sns
```

```python
bahsisler = sns.load_dataset('tips')
```

```python
bahsisler.head()
```

```python
sns.displot(bahsisler['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=bahsisler,kind='scatter')
```

```python
sns.pairplot(bahsisler)
```

```python
sns.pairplot(bahsisler, hue='sex')
```

```python
bahsisler.head()
```

```python
sns.pairplot(bahsisler, hue='smoker')
```

```python
sns.barplot(x='sex', y='total_bill', data=bahsisler)
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=bahsisler)
```

```python
bahsisler.corr()
```

```python
sns.heatmap(bahsisler.corr())
```

```python
sns.heatmap(bahsisler.corr(), annot=True)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

```python
g = sns.PairGrid(iris, hue='species')
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(bahsisler, col='time', row='smoker')
g = g.map(plt.hist, 'total_bill')
```

```python
g = sns.JointGrid(x='total_bill', y='tip', data=bahsisler)
g = g.plot(sns.regplot, sns.histplot)
```

```python
sns.countplot(x='day', data=bahsisler)
```
